```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Select multiple columns using a list
#labelencoding
columns_to_encode = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'native_country']
selected_data = adults1[columns_to_encode]
# Create an instance of LabelEncoder
label_encoder = LabelEncoder()
# Apply LabelEncoder to each column separately
encoded_datas = selected_data.apply(label_encoder.fit_transform)
```

```python
#Select multiple columns using a list
#one-hot encoding
columns_to_encodes = ['race','sex']
selected_data = adults1[columns_to_encodes]

# Create an instance of OneHotEncoder
one_hot_encoder = OneHotEncoder()

# Reshape the selected data to a 2D array
# -1 infers the number of rows based on the original shape
selected_data_2d = selected_data.values.reshape(-1, 1)

# Apply OneHotEncoder to the reshaped data
encoded_data = one_hot_encoder.fit_transform(selected_data_2d)
```

```python
# Concatenate the encoded data with the original dataset
from pandas import concat
import pandas as pd

# Convert the NumPy array to a DataFrame
encoded_data_df = pd.DataFrame(encoded_data.toarray())

encoded_data_result = concat([adults1.drop(columns_to_encode + columns_to_encodes, axis=1),
                       encoded_datas,
                       encoded_data_df], axis=1) # Concatenate DataFrames
```

```python
# Initiating the KaplanMeierFitter model
kmf = KaplanMeierFitter()
```

```python
# Handle missing values before fitting the model
T = encoded_data_result.age.dropna()  # Remove rows with missing values in 'age' column

# Fitting KaplanMeierFitter model on Time and Events for death
event_column_name = 'age'
kmf.fit(T, event_observed=encoded_data_result.loc[T.index, event_column_name])  # Use .loc to align event data with non-missing ages

# Time-line estimations plot
kmf.plot()
```
