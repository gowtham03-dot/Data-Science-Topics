```python
df=pd.get_dummies(Ship1)
df.head()
```

```python
#apriori algorithm
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
rules.sort_values('lift',ascending = False).head(10)
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

**Q1)** Lift : It is a measure of how much more the likely the two items are to be bought together and is calculated as the ratio of the observed support of the two items together if they were independent and it is important in the association rules because it helps us to identify rules that are truly interesting and not just due to chance.


**Q2)**Support: is an indication of how frequently an itemset appears in the dataset. It is the proportion of transactions that contain the itemset.
Support(X) = (Transactions containing X) / (Total Transactions)
Support(X->Y) = (Transactions containing X and Y) / (Total Transactions)
Confidence: indicates how likely item Y is purchased when item X is purchased, expressed as {X -> Y}. This is measured by the proportion of transactions with item X, in which item Y also appears.
Confidence(X->Y) = Support(X->Y) / Support(X)
